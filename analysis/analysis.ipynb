```python
data1 = data[(data["trip_distance"] < 100) & (data["trip_distance"] >= 0) & (data["tip_amount"] <= 600) & (data["tip_amount"] >= 0)]

fig, ax = plt.subplots()

ax.set_xlabel('trip distance')
ax.set_ylabel('tip amount')

ax.scatter(x=data1['trip_distance'], y=data1['tip_amount']);
```

```python
data2 = data[(data["passenger_count"] > 0) & (data["tip_amount"] <= 600) & (data["tip_amount"] >= 0)]

data2.plot.scatter(x='passenger_count', y='tip_amount').set_xticks(np.arange(0, 11, 1.0))
plt.show()
```
